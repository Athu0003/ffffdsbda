```python
import seaborn as sns
import matplotlib.pyplot as plt
#%matplotlib notebook
%matplotlib inline
```

```python
sns.lineplot(data=ds_heart[["Age","AHD"]], x="Age", y="AHD")
plt.show()
```

```python
sns.barplot(data=ds_heart[["Age","ChestPain"]], x="ChestPain", y="Age")
```

```python
x=ds_heart["ChestPain"].map({'typical':1,'asymptomatic':2,'nonanginal':3,'nontypical':4})
data=[]
for i in range(1,5):
  data.append(x.value_counts()[i])
print(data)
keys=['typical','asymptomatic','nonanginal','nontypical']
```

```python
palette_color = sns.color_palette('bright')

plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
```

```python
sns.histplot(data=ds_heart,x='Chol')
```

```python
plt.scatter(x='Age',y='RestBP', data=ds_heart, c='g') 
plt.scatter(x='Age',y='Chol', data=ds_heart, c='r', marker='+') 
plt.xlabel('Age')
plt.ylabel('Chol and RestBP') 
plt.legend(title='scatter_plot') 
plt.show()
```

```python
plt.figure() 
plt.legend(title='Scatter_plot')
sns.lmplot(x='Age', y='Chol', data=ds_heart, fit_reg=False)
```

```python
plt.plot(ds_heart["RestBP"],ds_heart["Age"])
plt.xlabel("Age")
plt.ylabel("AHD")
plt.show()
```

```python
sns.barplot(data=ds_heart[["Age","ChestPain"]], x="ChestPain", y="Age")
plt.bar(ds_heart["ChestPain"], ds_heart["Age"])
 
plt.xlabel("ChestPain")
plt.ylabel("Age")
plt.title("Bar Graph")
plt.show()
```
